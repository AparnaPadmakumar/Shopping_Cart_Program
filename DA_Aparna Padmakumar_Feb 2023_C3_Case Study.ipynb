```python
def print_items(store_items):
    print("----------ESSENTIAL ITEMS AVAILABLE---------------\n")
    print("%10s %10s %10s %10s"%("S.no","Item","Quantity","Cost/Item"))
    for name, data in store_items.items():
        print("%10d %10s %10d %10d"%(data["Srno"],name,data["stock"],data["cost"]))

def take_orders(store_items):
    cart=[]
    bill_items=[]
    total_item_cost = 0
    for name,data in store_items.items():
        while True:
            quantity=int(input("\nHow many packets of "+name+" are required? "))
            if quantity <= data['stock']:
                cart.append(quantity)
                total_item_cost= total_item_cost + quantity*data["cost"]
                data["stock"]=data["stock"]-quantity
                bill_items.append({"Srno":data["Srno"],"Name":name,"Quantity":quantity,"Total Cost":quantity*data["cost"]})
                break
            elif data["stock"]==0:
                print("Out of stock!!")
                break
            else:
                print("Only",data["stock"]," are left")
    return total_item_cost, bill_items

def print_bill(total_item_cost, bill_items, distance, person_name, address):
    print("\n--------------BILL--------------\n")
    print("%10s %10s %10s %10s"%("S.no","Item","Quantity","Total Cost"))
    for item in bill_items:
        print("%10d %10s %10s %10.2f"%(item["Srno"],item["Name"],item["Quantity"],item["Total Cost"]))

    print("Total Item Cost=",total_item_cost)
    if distance<=5:
        delivery_charge=10
    elif distance>5 and distance<=10:
        delivery_charge=12
    elif distance>10 and distance<=15:
        delivery_charge=14
    else:
        delivery_charge=20
    print("Delivery Charge= ",delivery_charge)
    print("Total Bill Amount=Total Item Cost + Delivery Charge=$ ",total_item_cost + delivery_charge)
    print("\nName= ",person_name)
    print("Address= ",address)

store_items={"Biscuits":{"Srno":1,"stock":5,"cost":2.50},
             "Cereals":{"Srno":2,"stock":10,"cost":3.50},
             "Chicken":{"Srno":3,"stock":20,"cost":13.30},
             "Oats":{"Srno":4,"stock":20,"cost":3.90},
             "Rice":{"Srno":5,"stock":30,"cost":4.50}}

print_items(store_items)

print("Fill your delivery details")
person_name = input("Enter your name: ")
address = input("Enter your address: ")
distance = float(input("Distance from Store 0-5/5-10/10-15/Above 15: "))

prev_total_item_cost=0
while True:
    
    total_item_cost, bill_items = take_orders(store_items)
    prev_total_item_cost=prev_total_item_cost+total_item_cost
    print_bill(total_item_cost, bill_items, distance, person_name, address)

    choice = input("\nDo you want to continue shopping? y/n: ")
    if choice == 'n':
        break
    elif choice == 'y':
        continue
    else:
        print("Invalid input. Exiting...")
        break
print("Total cost of all items purchased: $",prev_total_item_cost)
```
